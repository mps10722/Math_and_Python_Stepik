```python
def wisdom_multiplication(x, y, length_check = True):  #Функция расчета произведения по "Схеме мудреца"
    fst = str(100 - (100 - x) - (100 - y)) #Первая часть слагаемого
    sec = str((100 - x) * (100 - y))    #Вторая часть слагаемого (будем добавлять лидирующий 0, если 1) получилось однозначное число; 2) флаг length_check = 1)
    return int(fst + [sec, '0' + sec][length_check and len(sec) == 1]) #Вывод результата с или без лидирующего нуля
```

```python
def multiplication_check(x, y, length_check = True): #Функция проверки корректности работы "Схемы мудреца"
    return wisdom_multiplication(x, y, length_check) == x * y #Возвращае True, если Схема работает, иначе False
```

```python
def multiplication_check_list(start=10, stop=99, length_check = True): #Функция сбора статистики срабатывания "Схемы мудреца" на парах натуральных чисел, где каждое принимает значения от start до stop
    res = sum([int(multiplication_check(i, j, length_check)) for i in range(start, stop + 1) for j in range(start, stop + 1)]) #Подсчет количества пар чисел, где "Схема мудреца" сработала
    print('Правильных ответов: ' + str(res) + '\n' + 'Неправильных ответов: ' + str((stop - start + 1) ** 2 - res)) #Вывод результата (также экспертно посчитано количество пар чисел, где Схема не сработала)
```

```python
import pandas as pd
import numpy as np
```

```python
start = 10 #Начало диапазона натуральных чисел
stop = 100 #Конец диапазона натуральных чисел
matrix = [[str(y) + 'X' + str(x) for x in range(start, stop)] for y in range(start, stop)] #Матрица MxN, где в каждую ячейку запишем пару чисел, на которых будет проверяться "Схема мудреца"
```

```python
def mult_chck(st): #Функция присваивает цвет ячейке на пересечении чисел x и y в зависимости от результата проверки работы "Схемы мудреца"
    color = 'red' #По умолчанию цвет ячейки красный
    if multiplication_check(x = int(st[:st.index('X')]), y = int(st[st.index('X') + 1:]), length_check = False): #Если Схема работает и без добавления лидирующего нуля во вторую часть слагаемого Схемы,
        color = 'green' #окрасим ячейку зеленым
    elif multiplication_check(x = int(st[:st.index('X')]), y = int(st[st.index('X') + 1:]), length_check = True): #Если Схема работает только с добавлением лидирующего нуля во вторую часть слагаемого Схемы,
        color = 'brown' #окрасим ячейку коричневым
    return 'color: %s' % color #присвоим переменной color значение цвета (для описания стиля в css)
```

```python
df = pd.DataFrame(data = matrix, columns = range(start, stop), index = range(start, stop)) #Формируем DataFrame по матрице чисел
#df.style.applymap(mult_chck).render().split('\n') #Формируем файл стилей с раскрашенными ячейками для CSS
```
