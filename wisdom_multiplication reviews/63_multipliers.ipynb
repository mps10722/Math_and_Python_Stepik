```python
#версия python 3.7.1
#функция умножения по "схеме мудреца" с конкатенацией
def wisdom_multiplication(x, y, length_check = True):
    step_one = (100 - x) + (100 - y)
    step_two = 100 - step_one
    step_three = (100 - x) * (100 - y)
    
    if length_check:
        if len(str(step_three)) == 1:
            step_three = '0' + str(step_three)
    
    return str(step_two) + str(step_three)

#функция для проверки работы "схемы мудреца"
def color_multiplication(val):
    #функция принимает значение ячейки в виде строки с 2-мя множителями и знаком умножения между ними 
    x = int(val[:2]) #определяем множитель 1
    y = int(val[3:]) #определяем множитель 2
    s = int(wisdom_multiplication(x,y, length_check = False))
    #проверяем работает ли умножение по "схеме мудреца" с конкатенацией БЕЗ проверки длины step_three
    if s == x * y:
        color = 'green' #если работает, присваиваем ячейке зеленый цвет
        return 'color: %s' % color
    else:
        #если нет, то проверяем работает ли умножение по "схеме мудреца" с конкатенацией с проверкой длины step_three, т.е с добавлением при необходимости "0" 
        if int(wisdom_multiplication(x,y, length_check = True)) == x * y:
            color = 'blue' #если работает, присваиваем ячейке синий цвет
            return 'color: %s' % color
        else:
            #если и так умножение по "схеме мудреца" не работает, присваиваем ячейке красный цвет
            color = 'red'
            return 'color: %s' % color
```

```python
import pandas as pd

multipliers = [int(i) for i in range(10, 100)]

multipliers_two = {}
for x in range(10, 100):
    multipliers_two[x] = [str(x) + '*' + str(y) for y in range(10,100)]
    
df = pd.DataFrame(data = multipliers_two, index = multipliers)
pd.set_option('display.max_categories', 90)
pd.set_option('display.max_columns', 90)
pd.set_option('display.max_rows', 90)

df_new = df.style.applymap(color_multiplication)
with open('63_output.html', 'w') as file:
    file.write(df_new.render())
df_new
```
