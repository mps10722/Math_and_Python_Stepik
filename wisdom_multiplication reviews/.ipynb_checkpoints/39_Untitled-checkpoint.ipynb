```python
import pandas as pd
```

```python
#Копирую свои функции, которые делал в предыдущих заданиях
#wisdom_multiplication - перемножает числа по правилу мудреца с добавлением нуля или без, в зависимости от третьего параметра
#multiplication_check - проверяет, что перемножение получилось корректным
def wisdom_multiplication(x, y, length_check = True):
    part_1 = str(100 - ((100 - x) + (100 - y)))
    part_2 = str((100 - x) * (100 - y))
    if length_check:
        if len(part_2) == 1:
            part_2 = '0' + part_2
    return int(part_1 + part_2)
def multiplication_check(x, y, length_check = True):
    if (x*y) == wisdom_multiplication(x,y, length_check):
        return True
    else:
        return False
```

```python
#Создаем словарь, который будем заполнять строками вида X x Y, где X и Y - множители
#И создаем датафрейм из этого словаря
dict_multi = {}
for i in range(10, 100):
    dict_multi[i] = []
    for j in range(10, 100):        
        dict_multi[i].append(str(i) + 'x' + str(j))
df = pd.DataFrame(dict_multi)
df.index += 10
pd.set_option('display.max_rows', None, 'display.max_columns', None)
df
```

```python
#Создаем функцию, которая будет красить ячейки, в зависимости от того, правильный результат дает перемножение по мудрецу
#И проверяет это в двух случаях, требуется ли добавлять "0" или умножается и без этого
#Если умножение по мудрецу удалось, то красим в зеленый
#Если не удалось, то проверяем, можем получится, если добавлять ноль, если получилось, красим в желтый
#Если так и так не получается, то ничего не красим
def _color_green_yellow(val):
    first_num = int(val[0:2])
    second_num = int(val[3:5])
    if(multiplication_check(first_num, second_num, False)):
        return 'background-color: green'
    if(multiplication_check(first_num, second_num)):
        return 'background-color: yellow'
    return ''
```

```python
#Применяем стиль к датафрейму и сохраняем его в html файл
s = df.style.applymap(_color_green_yellow)
with open('index.html', 'w') as html:
    html.write(s.render())
s
```
