### Создаем таблицу 90 x 90 заполненную нулями, затем заполняем ее записями формата M x N

```python
tabel = [[0 for i in range(10, 100)] for j in range(10, 100)]

indexes = [i for i in range(10, 100)]

for i in range(10, 100):
    for j in range(10, 100):
        tabel[j - 10][i - 10] = f'{i}x{j}'

df = pd.DataFrame(tabel, index=indexes, columns=indexes)
```

### Используем функцию проверяющую возможность испоьзования "схемы мудреца"

```python
def multiplication_check(x, y, length_check = True):
    a = 100 - x
    b = 100 - y
    left = str(100 - a - b)
    right = '0' * (length_check and len(str(a * b)) == 1) + str(a * b)
    return int(left + right) == (x * y)
```

### Создадим списки из запесей M x N которые удовлетворяют схеме с/без проверки на длинну

```python
ls_1 = []
ls_2 = []
for i in range(10, 100):
    for j in range(10, 100):
        if multiplication_check(i, j, length_check=False):
            ls_1 += [f'{i}x{j}'] # список MxN удовлетворяющий БЕЗ проверки
        elif multiplication_check(i, j, length_check=True):
            ls_2 += [f'{i}x{j}'] # список MxN удовлетворяющий С проверкой
```

### Кра

```python
def color_tabel(val):
    if val in ls_1:
        color = 'green'
    elif val in ls_2:
        color = 'yellow'
    else:
        color = 'white'
    return f'background-color: {color}'
```
