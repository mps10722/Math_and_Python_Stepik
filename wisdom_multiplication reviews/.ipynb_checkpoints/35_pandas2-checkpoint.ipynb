```python
import numpy as np
import pandas as pd
def wisdom_multiplication(x, y, length_check=True):
    s1 = str(100 - (100 - x) - (100 - y))
    s2 = str((100 - x) * (100 - y))
    return s1 + '0' + s2 if length_check and len(s2) == 1 else s1 + s2
cols = []
ind = []
Leng = 90
ms = [0] * Leng
for i in range(10, 10 + Leng):
    cols += [i]
    ind += [i]
    ms[i - 10] = [0] * Leng
    for j in range(10, 10 + Leng):
        ms[i - 10][j - 10]  = str(i) + ' x ' + str(j)
    # ms.append(md[:])
# print(cols, ind, ms)
# A = [0] * 10
# for i in range(10): 
#     A[i] = [0] * 10
# print(A)
nda1 = [[1, 2, 3], [10, 20, 30]]
df4 = pd.DataFrame(ms, index=ind, columns = cols)
df4
```

```python
def font_color_mod(val):
    color_val = "black"
    val = val.split()
    a, b = int(val[0]), int(val[2])
    #     print(a * b, wisdom_multiplication(a, b, False), wisdom_multiplication(a, b))
    if a * b == int(wisdom_multiplication(a, b, False)):
        color_val = "red"
    elif a * b == int(wisdom_multiplication(a, b)):
        color_val = "blue"
    return f"color: {color_val}"
df_font_mod = df4.style.applymap(font_color_mod)
df_font_mod
```
