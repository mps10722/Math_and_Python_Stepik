```python
'''
Функция выдачи ответов
Вход: x, y - целые числа от 10 до 99 включительно
Выход: кортеж из трех целых чисел
    0 - реальный результат
    1 - результат без контроля цифры "0"
    2 - результат с контролем цифры "0"
'''
def my_test(x, y):
    temp_real = x * y
    d1 = str(x + y - 100)
    d2 = str((100 - x) * (100 - y))
    temp_test1 = int(d1 + d2)
    temp_test2 = int(d1 + d2.rjust(2, '0'))
    return (temp_real, temp_test1, temp_test2)
```

```python
start = 10
stop = 100

# Создание HTML файла
with open('control.html', mode='w', encoding='utf-8') as f:
    f.write('''
        <!DOCTYPE html>
        <html>
        <head>
            <meta charset="UTF-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Таблица умножения двузначных чисел от "Мудреца"</title>
            <style type="text/css">
                table {border-spacing: 0;}
                td {padding: 0;}
                .cell {display: inline-block; border: 1px solid gray; width: 55px; text-align: center;}
                .ans1 {background-color: green; color: white;}
                .ans2 {background-color: royalblue; color: white;}
            </style>
        </head>
        <body>
            <p>
                <h3>Обозначения:</h3>
                <div class="cell">10 x 10</div>
                <div style="display: inline-block;">
                    <span class="top">- искомое произведение</span><br>
                </div>
                <br><br>
                &nbsp;белый фон - неправильные ответы "Мудреца"<br>
                <span class="ans1">&nbsp;зелёный фон </span>&nbsp;- правильный ответ "Мудреца" без учёта "0"<br>
                <span class="ans2">&nbsp;синий фон </span>&nbsp;- правильный ответ "Мудреца" с учётом "0"
            </p>
            <table>
                <tbody>
    ''')
    
    itog = [0, 0, 0]  # список для подсчёта правильных и неправильных ответов
    # два цикла: перебор по стокам и столбцам
    for m in range(start, stop):
        f.write('<tr>')  # начало строки таблицы
        for n in range(start, stop):
            temp = my_test(m, n)  # получение ответов
            # проверка правильности ответа
            if temp[0] == temp[1]:
                answer = 'ans1'
                itog[1] += 1
            elif temp[0] == temp[2]:
                answer = 'ans2'
                itog[2] += 1
            else:
                answer = ''
                itog[0] += 1
            # запись ячейки таблицы
            f.write(f'<td><div class="cell {answer}">{m} x {n}</div></td>')

        f.write('</tr>\n')  # конец строки таблицы

    # запись итоговых результатов
    f.write(
        '<p><h3>Итог:</h3> '
        f'Bсего {sum(itog)} примеров, из них:<br>'
        f'&nbsp;{itog[0]} неправильные ответы;<br>'
        f'&nbsp;{itog[1]} правильный ответ без учёта "0";<br>'
        f'&nbsp;{itog[2]} правильный ответ с учётом "0".<br>'
        '</p>\n'
    )
    # "завершение" HTML файла
    f.write('''
                </tbody>
            </table>
        </body>
        </html>
    ''')
```
