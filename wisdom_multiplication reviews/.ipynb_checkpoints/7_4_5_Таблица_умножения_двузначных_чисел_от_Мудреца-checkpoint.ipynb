```python
import pandas as pd
import numpy as np
from IPython.display import display
#import sys
#print(sys.version_info)
# Python version 3.6.9

# Создаем таблицу с записями вида NxM
data_x = [ i for i in range(10,100)]
data_y = [ j for j in range(10,100)]
dan = np.zeros((90,90),dtype = np.unicode_)
for d in range(10,100):
    for k in range(10,100):   
        dan[d-10, k-10] = (str(d) + 'x' + str(k))    
df = pd.DataFrame(data=dan, index = data_y, columns=data_x, dtype=str)

# Создаем списки где работает "схема мудреца"
def multiplication_check_list(start = 10 , stop = 99, length_check = True):
    n = 0
    m = 0
    for x in range(start, stop + 1):
        
        for y in range(start, stop + 1):
          
            if multiplication_check(x, y, length_check):
                n +=1
            else:
                m += 1

spv = [] # список правильных пар (не требуется корректировка)

def multiplication_check(x, y, length_check):
    #z = simple_multiplication(x, y)
    z = wisdom_multiplication(x, y, length_check)
    if int(z) == int(x) * int(y):
        tmp = []
        tmp.append(x)
        tmp.append(y)
        spv.append(tmp)
        return True
    else:
        return False

def simple_multiplication(x, y):
    x1 = 100 - x
    y1 = 100 - y
    return ((100 -(x1 + y1))*100)+(x1*y1)
spvk = [] #  список пар с корректировкой
def wisdom_multiplication(x, y, length_check = True):
    x1 = 100 - x
    y1 = 100 - y
    z34 = 100 - (x1+y1)
    z12 = x1 * y1
    if length_check == True:
        if z12 <= 9:
            z = str(z34) + str(0) +str(z12)
            rez = int(z)
            tmp = []
            tmp.append(x)
            tmp.append(y)
            spvk.append(tmp)
            return rez
        else:       
            z = str(z34) + str(z12)
            rez = int(z)
            return rez
    else:       
        z = str(z34) + str(z12)
        rez = int(z)
        return rez
multiplication_check_list(10, 99, length_check = False)
multiplication_check_list(10, 99, length_check = True)

# визуально отмечаем цветом ячейки где работает схема мудреца
# зеленым цветом выделяем ячейки где работает
# желтым цветом выделяем ячейки  где работает (с корректировкой)
# оранжевым цветом выделяем ячейки где схема мудреца не работает
def create_colors(x):
    #copy df to new - original data are not changed
    df1 = x.copy()
    #select all values to default value - no color
    df1.loc[:,:] = 'background-color: '
    #overwrite values with green and red color
    for k in data_x:
        tmp = []
        for n in data_y:
            tmp.append(k)
            tmp.append(n)
            if tmp not in spv and tmp not in spvk:
                df1.loc[k,n] = 'background-color: darkorange'
    for i in range(len(spv)):
        x2 = spv[i][0]
        y2 = spv[i][1]    
        df1.loc[x2,y2] = 'background-color: green'
    for i in range(len(spvk)):
        x1 = spvk[i][0]
        y1 = spvk[i][1]
        df1.loc[x1,y1] = 'background-color: yellow'
    #return color df
    return df1      
print('зеленым цветом выделяем ячейки где работает схема мудреца')
print('желтым цветом выделяем ячейки  где работает схема мудреца (с корректировкой) ')
print('оранжевым цветом выделяем ячейки где схема мудреца не работает')
display(df.style.apply(create_colors, axis=None)) # отображаем таблицу в jupiter notebook
# выгружаем таблицу в формате html
with open("7_my.html","w") as fp:
   fp.write(df.style.apply(create_colors, axis=None).render())
```
