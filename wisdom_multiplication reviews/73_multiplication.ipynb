```python
# импортируем необходимые библиотеки
import pandas
```

```python
# делаем так, чтобы отображались все ячейки
pandas.set_option('display.max_rows', 500)
pandas.set_option('display.max_columns', 500)
pandas.set_option('display.width', 1000)
```

```python
def wisdom_multiplication(x, y, length_check=True):
    '''
    Функция выполняет перемножение двух двузначных чисел по методу мудреца
    '''
    a = 100 - x # вычитаем из 100 первое число
    b = 100 - y # вычитаем из 100 второе число
    
    h1 = str(100 - (a + b)) # вычитаем из 100 сумму этих чисел, получаем первые первые цифры результата 
    h2 = str(a * b) # получаем последние цифры результата
    
    if length_check and len(h2) < 2: # если производится проверка длины и длина последних цифр результата не 2
        h2 = '0' * (2 - len(h2)) + h2 # добавяем ведущие нули
    return int(h1 + h2) # конкатенируем результаты и возвращаем целое число
```

```python
# строим сетку, [10..99] x [10..99], в каждой ячейке которой записаны перемножаемые числа
grid = {x: ['{0} * {1}'.format(y, x) for y in range(10, 100)] for x in range(10, 100)}
```

```python
# получаем DataFrame, содержащий построенную сетку
data = pandas.DataFrame(grid, index=[x for x in range(10, 100)])
data
```

```python
def paint(line): # функция, которая определяет цвет клетки датафрейма
    # получаем x и y из строки вида 'x * y'
    x, _, y = line.split()
    x = int(x)
    y = int(y)
    
    # истинный результат умножения
    expected_result = x * y
    
    # перемножение по методу мудреца без проверки длины
    wisdom_result_no_check = wisdom_multiplication(x, y, length_check=False)
    
    # перемножение по методу мудреца с проверкой длины
    wisdom_result_check = wisdom_multiplication(x, y, length_check=True)
    
    if wisdom_result_check != expected_result:
        assert wisdom_result_no_check != expected_result
        # если результаты не сошлись даже с проверкой длины - красим ячейку в красный цвет
        return 'background-color: red'
    elif wisdom_result_no_check != expected_result:
        # если результаты сошлись с проверкой длины, но не сошлись без проверки - красим в синий
        return 'background-color: blue'
    else:
        # если сходится даже без проверки длины - красим в белый
        return 'background-color: white'
    

painted = data.style.applymap(paint) # красим сетку
```
