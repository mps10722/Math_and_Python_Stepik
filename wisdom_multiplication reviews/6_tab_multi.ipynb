```python
import numpy as np
import pandas as pd
```

```python
# снятие лимита отображаемых строк и столбцов
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

#Создание словаря tab_source исходных данных для создания DataFrame, 
# а также списка индексов ind
tab_source = {} 
ind = []
for i in range(10,100):
    line=[]
    for j in range(10,100):
        line.append(str(j)+'x'+str(i))
    tab_source[i] = line
    ind.append(i)
    
#Создание и вывод датафрейма    
df = pd.DataFrame(tab_source, index = ind)        
df
```

```python
# функция перемножения двух чисел по схеме мудреца, length_check - признак проверки на разделитель 0 внутри
def wisdom_multiplication(x, y, length_check = True):
    s1x,s1y = 100-x,100-y
    s1 = s1x + s1y
    s2 = 100 - s1
    p1 = s1x * s1y
    if length_check and p1<10:
        p1 = '0'+str(p1)
    res = str(s2)+str(p1)
    return res

# функция проверки справедливости схемы мудреца для двух чисел
def multiplication_check(x, y, length_check):
    return x*y==int(wisdom_multiplication(x, y, length_check))

# функция формирования стиля для ячейки
def color_multiplication_check(val):
    x,y =[int(s) for s in val.split('x')]
    # в случае справедливости схемы без учета проверки на разделитель 0 внутри ячейку красим в зеленый
    if multiplication_check(x,y,False): 
        return 'background-color: green' 
    # в случае справедливости схемы с учетом проверки на разделитель 0 внутри ячейку красим в желтый
    if multiplication_check(x,y,True):
        return 'background-color: yellow'
    return 'color: %s' % "black"

# стилизуем по цветам таблицу умножения
styled_tab = df.style.applymap(color_multiplication_check)
```

```python
# Сохраняем полученную отформатированную по цветам таблицу в файл
with open('out_file.html','w') as of:
    of.write(styled_tab.render())
```
