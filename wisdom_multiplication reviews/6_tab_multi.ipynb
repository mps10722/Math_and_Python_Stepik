```python
import numpy as np
import pandas as pd
```

```python
#      
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

#  tab_source     DataFrame, 
#     ind
tab_source = {} 
ind = []
for i in range(10,100):
    line=[]
    for j in range(10,100):
        line.append(str(j)+'x'+str(i))
    tab_source[i] = line
    ind.append(i)
    
#       
df = pd.DataFrame(tab_source, index = ind)        
df
```

```python
#       , length_check -     0 
def wisdom_multiplication(x, y, length_check = True):
    s1x,s1y = 100-x,100-y
    s1 = s1x + s1y
    s2 = 100 - s1
    p1 = s1x * s1y
    if length_check and p1<10:
        p1 = '0'+str(p1)
    res = str(s2)+str(p1)
    return res

#        
def multiplication_check(x, y, length_check):
    return x*y==int(wisdom_multiplication(x, y, length_check))

#     
def color_multiplication_check(val):
    x,y =[int(s) for s in val.split('x')]
    #          0     
    if multiplication_check(x,y,False): 
        return 'background-color: green' 
    #          0     
    if multiplication_check(x,y,True):
        return 'background-color: yellow'
    return 'color: %s' % "black"

#     
styled_tab = df.style.applymap(color_multiplication_check)
```

```python
#        
with open('out_file.html','w') as of:
    of.write(styled_tab.render())
```
